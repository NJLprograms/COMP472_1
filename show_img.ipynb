```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def print_img(file):
    df = pd.read_csv (file, header=None)

    # drop label
    del df[1024]

    # images are 32*32 
    images = [x.reshape(32,32) for x in df.to_numpy()]

    plt.imshow(images[0])
    plt.show()

print_img('Assig1-Dataset/train_2.csv')
```
